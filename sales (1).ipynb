```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('Diwali Sales Data.csv', encoding= 'unicode_escape')
```

```python
ax = sns.countplot(x='Gender', data=df)
ax.bar_label(ax.containers[0])
plt.show()
```

```python
ax = sns.countplot(data=df, x='Age Group', hue='Gender')
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
plt.show()
```

```python
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x = 'Age Group',y= 'Amount' ,data = sales_age)
```

```python
sales_state = df.groupby('State', as_index=False)['Orders'].sum().nlargest(8, 'Orders')
plt.figure(figsize=(15, 5))
sns.barplot(x='State', y='Orders', data=sales_state)
```

```python
sales_state = df.groupby('Product_Category', as_index=False)['Amount'].sum().nlargest(10, 'Amount')
plt.figure(figsize=(20, 5))  # Smaller and neat size
sns.barplot(x='Product_Category', y='Amount', data=sales_state)
plt.show()
```
